## If we have more than 3 feature and we can not draw 4d digram so to understand that features we can do multivariate analysis
## by using pairplot from seaborn we can see the behaiour of features(more than 3)

## In pairplot on y axis and x axis we have same feature 
## every features will be compare with the other features 
## By using those digrams we can get the knowledge of correlation
## if one feature is increaing and another feature is also increaing the we can say that there is poistive corelation
 
## if one feature is decreasing and another feature is also idecreaing the we can say that there is negative corelation

## if one is not affecting the other then we can say that corelation is zero
## Value will always be betwwen -1 to 1
##  from the given graph we can know the behaviour of the data and can also find which algorithm we can use on that data

## in pairplot catagorical features will not come only numeric varaible will be plot
