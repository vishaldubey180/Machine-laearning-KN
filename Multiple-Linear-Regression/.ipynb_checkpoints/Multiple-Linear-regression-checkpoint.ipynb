# Multiple Linear Regression

```python
# Importing the libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot
```

```python
# importing the dataset
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df
```

```python
## here we can see that our profit is dependent on multiple fectors this is why it is called multiple linear regression
```

```python
x=df.iloc[:,:-1] # Except last column select all the column (independent varaiable)
```

```python
y=df.iloc[:,4] # select only last column( dependent varaiable)
```

```python
x
```

```python
y
```

```python
# in column we can see that there is state where column is state which is catagorical 
```

```python
# machine learning can not work with text data 
```

```python
# so we have to convert that into numeric form using one hot encoding
```

```python
# first we will use pandas get_dummies method to convert catagorical varaiable into dummy variable
```

```python
dummy_states = pd.get_dummies(x['State'])
```

```python
dummy_states
```

```python
# we can avoid dummy varaiable trap by droping one column 
```

```python
dummy_states = pd.get_dummies(x['State'],drop_first=True) # it will remove first column
```

```python
# as we have created dummy varaiable for state column so we can drop state column from x dataset
```

```python
x = x.drop('State',axis=1)
```

```python
x
```

```python
# Now we will add dummy varaiable to our dataframe
```

```python
x = pd.concat([x,dummy_states],axis=1)
```

```python
x
```

```python
# Splitting the dataset into training and testing dataet
```

```python
from sklearn .model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# fitting multiple linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
a = model.predict(X_test)
```

```python
# We can also use R2 method
```

R-squared is a statistical measure that represents the goodness of fit of a regression model. The ideal value for r-square is 1. The closer the value of r-square to 1, the better is the model fitted.

R-square is a comparison of residual sum of squares (SSres) with total sum of squares(SStot). Total sum of squares is calculated by summation of squares of perpendicular distance between data points and the average line.


Residual sum of squares in calculated by the summation of squares of perpendicular distance between data points and the best fitted line.

R square is calculated by using the following formula :

Where SSres is the residual sum of squares and SStot is the total sum of squares.

The goodness of fit of regression models can be analyzed on the basis of R-square method. The more the value of r-square near to 1, the better is the model.

Note : The value of R-square can also be negative when the models fitted is worse than the average fitted model.

```python
# Let's use r2
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test,a)
```
